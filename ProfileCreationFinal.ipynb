```python
def generate_profiles(num_profiles):
    profiles = []
    for i in range(num_profiles):
        selected_sector = random.choice(list(sectors.keys()))
        skills_list = sectors[selected_sector]['skills']
        interests_list = sectors[selected_sector]['interests']
        industries_list = sectors[selected_sector]['industries']

        profile = {
            'profile_id': i + 1,
            'name': f'Person {i + 1}',
            'sector': selected_sector,
            'skills': ', '.join(random.sample(skills_list, random.randint(2, 5))),
            'interests': ', '.join(random.sample(interests_list, random.randint(2, 4))),
            'current_industry': random.choice(industries_list)
        }
        profiles.append(profile)
    return profiles
```

```python
alumni_profiles = generate_profiles(2000)
student_profiles = generate_profiles(100)
```

```python
alumni_df = pd.DataFrame(alumni_profiles)
student_df = pd.DataFrame(student_profiles)
```
